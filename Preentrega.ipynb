# Visualización de los primeros registros

## Incumplimiento de préstamo

En este caso, teniendo en cuenta los datos del cliente a los cuales tenemos acceso, se desea predecir si cumplirá o no con el abono de su préstamos.

Para ello, las variables más interesantes son:

- loan_amount: cantidad prestada
- rate_of_interest: interés del préstamo
- loan_type: tipo de préstamo
- loan_purpose: propósito del préstamo
- credit_score: puntaje crediticio
- status: si el préstamo fue incumplido (1) o no (0)

## Detección de fraude en transacciones con tarjetas de crédito

En este caso se debea, a partir de los datos del cliente y comercio, se desea determinar si una transacción es o no fraudulenta.

Para ello, las variables más interesantes son:

- User_ID: ID del usuario
- Transaction_Amount: monto de la transacción
- Transaction_Type: tipo de transacción
- Device_Type: tipo de dispositivo
- Transaction_Distance: distancia de la transacción
- Risk_Score: puntaje de riesgo calculado para la transacción
- Daily_Transaction_Count: cantidad de transacciones realizadas en el día
- Avg_Transaction_Amount_7d: promedio de transacciones por semana
- Is_Weekend: indica si las transacciones fueron realizadas un fin de semana
- Fraud_Label: etiqueta que indica si es fraude (1) o no (0)
